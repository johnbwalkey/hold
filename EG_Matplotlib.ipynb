```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 50) # linear space
y = np.sin(x)
z = np.cos(x)
```

The term sin(x) refers to the sine of ùë•
a trigonometric function that relates the angle ùë•
to the ratio of the length of the opposite side to the hypotenuse in a right-angled triangle.

A basic way to create a plot

```python
fig, ax= plt.subplots()
ax.plot(x,y)
plt.show()
```

add a color and line style

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'r--')
plt.show()
```

more readable color - marker - linestyle
SEE examples (https://matplotlib.org/stable/gallery/index.html)

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='b', linestyle='-', marker='o')
plt.show()
```

color can be hexadecimal & define size of figure

```python
fig, ax= plt.subplots(figsize = (6,3)) # width then height
ax.plot(x, y, color='#7a21bf', linestyle='-', marker='o')
plt.show()
```

change line width and add label

```python
fig, ax= plt.subplots()
ax.plot(x, y, color='g', linewidth= 8, label = 'sin (x)')
ax.legend(loc = 'lower left')
plt.show()
```

```python
# 
fig, ax = plt.subplots()
ax.plot(x, y, color='y', linewidth= 8, label = 'sin (x)')
ax.set_yticks([-1, 0, 1])
ax.legend (loc = 'upper center')
plt.show()
```

```python
# add labels to x - y axis and title above
fig, ax = plt.subplots()
ax.plot(x, y, color='y', linewidth= 8, label = 'sin (x)')
ax.set_yticks(np.linspace(-1, 1, 7))
ax.legend (loc = 'lower right')
plt.ylabel('y-value')
plt.xlabel('x-value')
plt.title('Sin(x) function')
plt.show()
```

```python
# plot 2 series on the same graph
fig, ax = plt.subplots()
ax.plot(x, y, color='r', linewidth= 8, label = 'sin (x)')
ax.plot(x, z, color='b', linewidth= 8, label = 'cos (x)')
ax.set_yticks(np.linspace(-1, 1, 10))
ax.legend (loc = 'lower left')
plt.ylabel('y-value')
plt.xlabel('x-value')
plt.title('Sin(x) and Cos(x) functions')
plt.show()
```

```python
# plot several graphs in one figure
fig, ax = plt.subplots(2, 2, figsize = (10, 10))
ax[0,0].plot(x, y, color='r', linewidth= 8, label = 'sin (x)')
ax[0,1].plot(x, z, color='b', linewidth= 8, label = 'cos (x)')
ax[1,0].plot(x, y, color='g', linewidth= 8, label = 'sin (x)')
ax[1,1].plot(x, z, color='y', linewidth= 8, label = 'cos (x)')
plt.show()
# can right click on image and copy or save as image
```

## Other types of plots

```python
x_values =[1, 2, 3, 4, 5, 6]
y_values = [1, 4, 9, 16, 25, 36]
fig, ax = plt.subplots()
ax.plot(x_values, y_values, color='r', linewidth= 8, label = 'y = x^2')
ax.set_yticks(np.linspace(0, 36, 7))
ax.legend (loc = 'lower right')
plt.ylabel('y-value')
plt.xlabel('x-value')
plt.title('y = x^2 function')
```

## Bar Plots

```python
age=['16-20', '21-25', '26-30', '31-35', '36-40']
height=[160, 165, 170, 175, 180]
fig, ax = plt.subplots()
ax.bar(age, height, color='r', linewidth= 8, label = 'height')
ax.set_yticks(np.linspace(0, 200, 10))

plt.ylabel('height (cm)')
plt.xlabel('age group')
plt.title('Height vs Age group')
plt.show()
```

```python
# histogram
np.random.seed(0)
data = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(data, bins=30, color='r', label = 'height')
ax.set_yticks(np.linspace(0, 200, 10))

plt.ylabel('height (cm)')
plt.xlabel('age group')
plt.title('Height vs Age group')
plt.show()
```
