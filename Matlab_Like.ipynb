```python
# ===== MATLAB-LIKE PYTHON NOTEBOOK SETUP =====

import numpy as np
import matplotlib.pyplot as plt

# Plotting defaults (MATLAB-ish)
plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["axes.grid"] = True
plt.rcParams["lines.linewidth"] = 2

# Inline plots (safe to include)
%matplotlib inline

# Numpy display similar to MATLAB
np.set_printoptions(precision=4, suppress=True)

# MATLAB-style convenience aliases
linspace = np.linspace
zeros    = np.zeros
ones     = np.ones
rand     = np.random.rand
randn    = np.random.randn

# Optional MATLAB-like plot() wrapper
def plot(x, y=None, style='-'):
    if y is None:
        plt.plot(x, style)
    else:
        plt.plot(x, y, style)
    plt.show()

print("MATLAB-like Python environment ready.")
```

```python
x = linspace(0, 10, 200)
y = np.sin(x)

plot(x, y, 'r-')
```
